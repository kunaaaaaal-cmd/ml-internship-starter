# Week 1 Research Question

## Selected Lane

CTR / Engagement Opportunity Scoring


## Search Question

Which webpages have the greatest opportunity to improve their click-through rate (CTR), and which page characteristics are associated with low CTR despite receiving many impressions?


## Unit of Analysis

Each webpage in the dataset.

## Expected Output

A ranked list of webpages with the highest opportunity to improve CTR and the factors contributing to their performance.

## Action Someone Could Take

SEO specialists and content writers can prioritize webpages with high impressions but low CTR. They can improve page titles, meta descriptions, and content to increase user engagement and clicks.

## Cost of a Wrong Recommendation

If the recommendations are inaccurate, teams may spend time optimizing pages with little improvement potential while overlooking pages that could significantly increase website traffic.

## Why Machine Learning Helps

Machine learning can analyze multiple page-level features simultaneously and identify hidden patterns that are difficult to detect manually. It helps prioritize webpages with the highest potential for improving CTR.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
# Supporting statistics for my chosen lane

print("Total Pages:", len(df))
print("Average CTR:", round(df["ctr"].mean(), 4))
print("Average Impressions (90d):", round(df["impressions_90d"].mean(), 2))
print("Average Position:", round(df["avg_position"].mean(), 2))

print("\nTrend Direction Distribution:")
print(df["trend_direction"].value_counts())
```

## Supporting Evidence

The starter dataset contains **30,000 webpages** with **44 features**, providing sufficient data for CTR analysis.

The calculated statistics show the average CTR, average impressions over the last 90 days, and average search position. The trend direction distribution further demonstrates that webpages perform differently over time, making it possible to identify pages with high improvement potential.
